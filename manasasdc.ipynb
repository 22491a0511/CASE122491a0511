```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

harsha=pd.read_csv('energy.csv')
print(harsha)

#line chart
#andhra,karnataka
srh=harsha[harsha['State']=='Andhra']
rcb=harsha[harsha['State']=='Karnataka']

fig,diag=plt.subplots(2)
#data,consumption
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])
```

```python
import pandas as el
import matplotlib.pyplot as plt
import seaborn as cse
mahesh=el.read_csv('heart.csv')
print(mahesh)

#scatter plot
cse.scatterplot(x='age',y='bp',data=mahesh,hue='gender')
plt.title('heart attack')
plt.xlabel('age')
plt.ylabel('BP')
plt.show()

#line chart
cse.lineplot(x='age',y='cholestrol',data=mahesh)
plt.show()

#box plot
color=['purple','green']
cse.boxplot(x='target',y='value',data=mahesh,palette=color)
plt.show()

#count plot
color=['skyblue','pink']
cse.countplot(x='target',data=mahesh,palette=color)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

yamini=pd.read_csv('energy.csv')
print(yamini)

#LINE CHART ANDHRA KARNATAKA
srh=yamini[yamini['State']=='Andhra']
rcb=yamini[yamini['State']=='Karnataka']

fig,diag=plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])

cse4='Kerala'
cse5='Delhi'
tdp=[cse4,cse5]

srh=yamini[yamini['State']=='Kerala']['Consumption'].sum()
rcb=yamini[yamini['State']=='Kerala']['Consumption'].sum()
ysr=[srh,rcb]

fig,dig=plt.subplots(2)
diag[0].pie(ysr,labels=tdp)
```
